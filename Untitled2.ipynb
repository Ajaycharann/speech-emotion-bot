```python
from google.colab import output
import base64, io
from pydub import AudioSegment

record_audio_js = """
const sleep  = time => new Promise(resolve => setTimeout(resolve, time))
const b2text = blob => new Promise(resolve => {
  const reader = new FileReader()
  reader.onloadend = e => resolve(reader.result)
  reader.readAsDataURL(blob)
})

async function record(sec){
  stream = await navigator.mediaDevices.getUserMedia({audio:true})
  recorder = new MediaRecorder(stream)
  chunks = []
  recorder.ondataavailable = e => chunks.push(e.data)
  recorder.start()
  await sleep(sec*1000)
  recorder.stop()
  await new Promise(resolve => recorder.onstop=resolve)
  blob = new Blob(chunks, {type: 'audio/webm'})
  return await b2text(blob)
}
record(5)  // record 5 seconds
"""

def record_audio():
    audio_bytes = output.eval_js(record_audio_js)
    audio_str = audio_bytes.split(',')[1]
    audio_data = base64.b64decode(audio_str)

    # Convert webm → wav
    audio = AudioSegment.from_file(io.BytesIO(audio_data), format="webm")
    audio.export("user_audio.wav", format="wav")
    return "user_audio.wav"

audio_path = record_audio()
print("🎤 Saved audio:", audio_path)
```

```python
from transformers import AutoFeatureExtractor, AutoModelForAudioClassification
import torch, librosa

emotion_model = "superb/wav2vec2-base-superb-er"
extractor = AutoFeatureExtractor.from_pretrained(emotion_model)
emo_model = AutoModelForAudioClassification.from_pretrained(emotion_model)

def detect_emotion(audio_path):
    speech, sr = librosa.load(audio_path, sr=16000)
    inputs = extractor(speech, sampling_rate=16000, return_tensors="pt", padding=True)
    with torch.no_grad():
        logits = emo_model(**inputs).logits
    emotion_id = torch.argmax(logits, dim=-1).item()
    return emo_model.config.id2label[emotion_id]

emotion = detect_emotion(audio_path)
print("🧠 Detected Emotion:", emotion)
```

```python
from transformers import pipeline
asr = pipeline("automatic-speech-recognition", model="openai/whisper-small")
user_text = asr(audio_path)["text"]
print("🗣 You said:", user_text)
```

```python
from google.colab import output
import base64, io
from pydub import AudioSegment
import torch, librosa
from transformers import AutoFeatureExtractor, AutoModelForAudioClassification, pipeline
from gtts import gTTS
from IPython.display import Audio

# ---------------- 🎤 Step 1: Record Audio ----------------
record_audio_js = """
const sleep  = time => new Promise(resolve => setTimeout(resolve, time))
const b2text = blob => new Promise(resolve => {
  const reader = new FileReader()
  reader.onloadend = e => resolve(reader.result)
  reader.readAsDataURL(blob)
})

async function record(sec){
  stream = await navigator.mediaDevices.getUserMedia({audio:true})
  recorder = new MediaRecorder(stream)
  chunks = []
  recorder.ondataavailable = e => chunks.push(e.data)
  recorder.start()
  await sleep(sec*1000)
  recorder.stop()
  await new Promise(resolve => recorder.onstop=resolve)
  blob = new Blob(chunks, {type: 'audio/webm'})
  return await b2text(blob)
}
record(5)  // record 5 seconds
"""

def record_audio():
    audio_bytes = output.eval_js(record_audio_js)
    audio_str = audio_bytes.split(',')[1]
    audio_data = base64.b64decode(audio_str)

    # Convert webm → wav
    audio = AudioSegment.from_file(io.BytesIO(audio_data), format="webm")
    audio.export("user_audio.wav", format="wav")
    return "user_audio.wav"

# ---------------- 🧠 Step 2: Emotion Detection ----------------
emotion_model = "superb/wav2vec2-base-superb-er"
extractor = AutoFeatureExtractor.from_pretrained(emotion_model)
emo_model = AutoModelForAudioClassification.from_pretrained(emotion_model)

def detect_emotion(audio_path):
    speech, sr = librosa.load(audio_path, sr=16000)
    inputs = extractor(speech, sampling_rate=16000, return_tensors="pt", padding=True)
    with torch.no_grad():
        logits = emo_model(**inputs).logits
    emotion_id = torch.argmax(logits, dim=-1).item()
    return emo_model.config.id2label[emotion_id]

# ---------------- 📝 Step 3: Speech-to-Text ----------------
asr = pipeline("automatic-speech-recognition", model="openai/whisper-small")

# ---------------- 🤖 Step 4: Chatbot ----------------
chatbot = pipeline("text-generation", model="microsoft/DialoGPT-medium")

def generate_response(user_text, emotion):
    prompt = f"The user is feeling {emotion}. Respond empathetically to their message: '{user_text}'"
    result = chatbot(prompt,
                     max_new_tokens=80,
                     pad_token_id=50256,
                     truncation=True)
    reply = result[0]['generated_text'][len(prompt):].strip()
    return reply

# ---------------- 🔊 Step 5: Text-to-Speech ----------------
def speak_text(text):
    tts = gTTS(text=text, lang="en")
    tts.save("response.mp3")
    return Audio("response.mp3", autoplay=True)

# ---------------- 🚀 Run Full Pipeline ----------------
audio_path = record_audio()                  # Record
emotion = detect_emotion(audio_path)         # Detect emotion
user_text = asr(audio_path)["text"]          # Transcribe
reply = generate_response(user_text, emotion)  # Chatbot reply

print(f"🗣 You said: {user_text}")
print(f"🧠 Emotion: {emotion}")
print(f"🤖 Bot: {reply}")
speak_text(reply)                            # Speak reply
```
